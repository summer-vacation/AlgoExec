# 链表反转

三个指针的应用：原列表的头指针、新列表的头指针、原列表的头指针的第二个指针(第二指针)

linked-list-reverse.png

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

root, p1, p2, p3, p4, p5 = (
    Node(0), Node(1), Node(2), Node(3), Node(4), Node(5))

root.next, p1.next, p2.next, p3.next, p4.next = (
    p1, p2, p3, p4, p5 )
```

```python
def print_list(root):
    iter_node = root
    while iter_node:
        print(iter_node.data, end=",")
        iter_node = iter_node.next
print_list(root)
```

```python
def reverse_list(root):
    """ 链表反转 """
    if not root or not root.next:
        return root
    
    # 指向新的首节点
    new_root = None
    curr_node = root
    while curr_node:
        # 保存当前节点的下一个节点
        next_node = curr_node.next   
        # 让当前节点的下一个节点变成新列表的root
        curr_node.next = new_root
        new_root = curr_node
        # 当前节点往下遍历
        curr_node = next_node
    return new_root
```

```python
print_list(reverse_list(root))
```
